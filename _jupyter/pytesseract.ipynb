```python
from PIL import Image
import pytesseract

# 실행파일 위치지정
pytesseract.pytesseract.tesseract_cmd = "C:/Program Files/Tesseract-OCR/tesseract.exe"
```

```python
imgfile = Image.open("image.png")
```

```python
imgfile
```

```python
text = pytesseract.image_to_string(imgfile)
```

```python
print(text)
```

```python
imgfile2 = Image.open("image2.png")
```

```python
imgfile2
```

```python
text2 = pytesseract.image_to_string(imgfile2, lang='kor')
```

```python
print(text2)
```

```python
print(text2)
```

```python
with open("result.txt", "w", encoding="utf8") as f:
    f.write(text2)
    f.write("\n\n\n")
    f.write(text2.replace(" ",""))
    
```
